```python
import pandas as pd
import numpy as np
import random
```

```python
data = pd.read_csv('car_data.csv')

#chèn thêm brand 

row, col = data.shape
Name = list(data['Name'])
brand = []
year = set(data['Year of manufacture'])

for i in range(row):
    brand.append(Name[i].split()[0])
    
brand = list(brand)
data['Brand'] = brand

#tạo df cho danh mục brand, giá tiền trung bình 
 
analistData = pd.DataFrame(columns=['Brand', 'Averange Price', 'Value'])
analistData['Brand'] = list(set(brand))
averange = []
value = []
sumNum = 0
AvgPrice = 0

for i in range(len(set(brand))):
    d = data[data['Brand'] == str(brand[i])]
    d.to_csv()
    listPrice = list(d['Price'])
    sumNum = 0
    AvgPrice = 0
    for j in range(len(listPrice)):
        AvgPrice = listPrice[j] + AvgPrice
    averange.append(AvgPrice/len(listPrice))
    value.append(len(listPrice))
 
analistData['Averange Price'] = averange
analistData['Value'] = value

analistData
```

```python
#tạo biểu đồ phân tích theo giá trung bình

analistData['Averange Price'].plot.bar(xlabel = 'Brand', ylabel = 'Averange Price')
```

```python
#tạo  biểu đồ phân tích theo số sản phẩm đăng bán

analistData['Value'].plot.bar(xlabel = 'Brand', ylabel = 'Value')
```
