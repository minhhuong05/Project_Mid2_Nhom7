```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


all_info = []
url = 'https://bonbanh.com/oto-cu-da-qua-su-dung'
link_list = []
link_list.append(url)

#Link truy cập vào từng trang của website
for i in range(2, 1481):
    new_link = url + f'/page,{i}'
    link_list.append(new_link)

for j in range(len(link_list)):
    proxy = {
    'http': "http://user-luongquynhchi118398-port-10000:1thyod6jqy@pr.zeusproxy.com:8888",
    'https': "http://user-luongquynhchi118398-port-10000:1thyod6jqy@pr.zeusproxy.com:8888"
    }

    response = requests.get(link_list[j], proxies=proxy)
    soup = BeautifulSoup(response.text, 'html.parser')        
    cars = soup.find_all("li", class_=["car-item row1", "car-item row2"] )

    for car in cars:
        model = car.find("div", class_="cb2_02").text
        link_car = car.find('a')
        link_car = link_car.get('href')


        split_name = model.replace('-', ' ').split()
        car_brand = split_name[0]
        year_of_manufacture = split_name[-1]
        if year_of_manufacture.isdigit():
            year_of_manufacture = split_name[-1]
        else:
            year_of_manufacture = 'No Result'

        price = car.find("div", class_="cb3").text

        split_price = price.split()
        if 'Tỷ' in split_price:
            if 'Triệu' in split_price:
                new_price = int(split_price[split_price.index('Tỷ') - 1]) * 1000 + int(split_price[split_price.index('Tỷ') + 1])
            else:
                new_price = int(split_price[split_price.index('Tỷ') - 1])
        elif 'Triệu' in split_price:
            new_price = int(split_price[split_price.index('Triệu') - 1])
        else:
            new_price = 'No Result'

        descript = car.find("div", class_="cb6_02")
        specifications = descript.contents[0]
        descriptive = descript.find("p").get_text()

        contact = car.find("div", class_="cb7")
        trader = contact.contents[1].text.strip()
        location = contact.contents[3].text.strip()
        phone_number = contact.contents[-1].strip()

        
        car_img = car.find('div', class_="cb5")
        car_img = car.find('img')
        car_img = car_img.get('src') 
        if 'https' not in car_img:
            car_img = 'No Picture'
        all_info.append([car_brand, model, year_of_manufacture, new_price, price, trader, location, phone_number, specifications, descriptive, link_car, car_img])
df = pd.DataFrame(all_info, columns=['Brand', 'Model', 'Year of manufacture', 'Gia so', 'Gia chu', 'Lien he', 'Dia chi', 'SDT', 'Ki thuat', 'Mo ta','Link xe', 'img'])
df.to_csv('car_data')
#'Brand', 'Model', 'Year of manufacture', 'Gia so', 'Gia chu', 'Lien he', 'Dia chi', 'Phone number', 'Ki thuat', 'Mo ta'
```
